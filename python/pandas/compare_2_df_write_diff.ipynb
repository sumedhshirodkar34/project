```python
import pandas as pd
import csv
```

```python
df7 = pd.read_csv('game_1.csv')
df7.columns = df7.columns.str.lower()
df7
```

```python
df8 = pd.read_csv('game_5.dat',sep = '|')
df8
```

```python
comparison_df_mix = df7.merge(df8,indicator=True, how='outer')
```

```python
comparison_df_mix
```

```python
comparison_df_filter = comparison_df_mix.loc[lambda x : x['_merge']!='both']
```

```python
comparison_df_filter
```

```python
df_not_matched = pd.DataFrame(columns = ['ID', 'Coulmn_name', 'Value'])
```

```python
for name, group in comparison_df_filter.groupby(['game_id','season']):
    for i in range(len(group)) :
        if(i < len(group)-1):
            for j in range(len(group.columns)) :
                if(j < len(group.columns) -1):
                    if(group.iloc[i, j] != group.iloc[i+1, j]):                        
                        df_not_matched = df_not_matched.append({'ID' : name, 'Coulmn_name' : group.columns[j], 'Value' : (group.iloc[i, j],group.iloc[i+1, j])}, ignore_index = True)
```

```python
df_not_matched.to_csv('notMatched_final_2.csv')
df_not_matched
```
